## Write a NWB file using NWBIO class from Neo

### Define the intrinsic structure in the different Neo object

```python
# Define Neo blocks
blk = Block(name='The block', session_start_time=datetime.now())

num_seg = 3  # number of segments
num_chan = 4  # number of channels

for ind in range(num_seg):  # number of Segments
    seg = Segment(index=ind)
    seg.block = blk
    blk.segments.append(seg)

for seg in blk.segments:  # AnalogSignal objects

    # 3 Neo AnalogSignals
    a = AnalogSignal(np.random.randn(44, num_chan) * pq.nA,
                                 sampling_rate=10 * pq.kHz,
                                 t_start=50 * pq.ms)
        
    b = AnalogSignal(np.random.randn(61, num_chan) * pq.nA,
                                 sampling_rate=3 * pq.kHz,
                                 t_start=0 * pq.ms)
        
    c = AnalogSignal(np.random.randn(10, num_chan) * pq.nA,
                                 sampling_rate=1 * pq.kHz,
                                 t_start=6 * pq.ms)

    seg.analogsignals.append(a)
    seg.analogsignals.append(b)
    seg.analogsignals.append(c)
        
    a.segment = seg
    b.segment = seg
    c.segment = seg
```

```python
blk
```

### Write the NWB file

```python
test_create_NWB_file_name = "test_Create_NWB_File.nwb"
iow = NWBIO(filename=test_create_NWB_file_name, mode='w')
iow.write_all_blocks([blk])
print("iow = ", iow)
```
